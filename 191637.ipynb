```python
#scatter plot
price = df["Price"]
price=list(price)

distance=df["Distance"]
distance=list(distance)
```

```python
fig=plt.figure(figsize = (7,7))

#line plot
plt.plot(distance, price, linestyle="", marker="*", markerfacecolor = "orange", markeredgecolor="black")

plt.xlabel("Distance")
plt.ylabel("price")
plt.title("distance vs price")
plt.show()
```
